# Algorithm Usage

### Helper Functions

1. **`run_model(data, num_chains = 8, p = 5, MCMC_steps=1000, hyperparam_steps=50, runtime = True)`**
    - **Returns**: Fitted model
        - The dataframe index needs to be reset as an integer.
        - Choose a high number for `MCMC_steps` to allow for a burn-in period.
        - `p` is our lag window

2. **`post_dep(model, num_samples)`**
    - **Returns**: Pairwise posterior dependence matrices (cluster probabilities)
        - The array is of size `NUM_SAMPLES x NUM_CHAINS x LEN(DF) x LEN(DF)`

3. **`clustering(post_probs, threshold = 0.75)`**
    - **Returns**: Clusters
        - If time series `i` and `k` are dependent in `threshold%` of samples, they are clustered together.
        - Averages over `NUM_SAMPLES` and `NUM_CHAINS`.

4. **`return_ari(true_labels, predicted_clusters)`**
    - **Returns**: Adjusted Rand Index (ARI) if ground truth is available

### Data Generation (Sin Waves)

In this section, we generate toy data using sine waves to demonstrate the usage of the TRCRP algorithm. Sine waves are chosen because they effectively replicate real-world cyclical and variable behavior, making them ideal for showcasing the algorithm's ability to capture patterns in complex data.

It's important to note that this toy data has no discernible patternâ€”it is neither straightforward nor intuitive to determine how many clusters exist within the data, nor what criteria make signals similar enough to be grouped together. Additionally, the data is indexed by time, and each observation is dependent on the previous one.

```python
import experiments.data_generation as data_gen
importlib.reload(data_gen)

df = data_gen.generate_sine_wave_data(frequency_noise=0.1, random_state=42, use_colors = False)
df.head()
```

## Run the Algorithm

*helper_functions.py* contains more details on the functions + fitting of the model

```python
data = df.iloc[1:]
labels = df.iloc[0].values

print(data.shape) # NUM_OBSERVATIONS X NUM_TIMESEREIES
```

### Fit the model

First, we fit the model with the data using the run_model() function.

Briefly, the parameters are:
- num_chains: increases robustness but also sampling time (needs to be optimized for parallelization)
- p: controls the impact of previous observations
- MCMC steps: inference iterations to approach convergence 
- hyperparam_steps: hyperparameter optimization; the model's ability to fine tune itself

```python
# fit the model
model = helpers.run_model(data)
```

### Sample from the Posterior

We sample from the posterior, which is crucial in estimating cluster assignments and identifying dependencies between signals. We can think of the posterior as a version of the model that has been fit to the data. This step is how we infer the underlying structure of the data. 

```python
# sample from the posterior (estimate clusters)
dep_matrices = helpers.post_dep(model, 20)
```

### Analysis: Compare Predicted v True Clusters

When we run *helpers.clustering()*, each set of numbers inside the curly brackets {} represents a cluster, where the numbers are the indices of the signals that belong to that cluster. For example, the first cluster contains signals 0, 5, 10, 15, and so on. This indicates that these signals exhibit similar temporal behaviors, which the model has identified and grouped together

```python
predicted_clusters = helpers.clustering(dep_matrices, threshold = 0.75)
predicted_clusters
```

To evaluate the clustering performance, we use the helpers.return_ari(labels, predicted_clusters) function. This function calculates the Adjusted Rand Index (ARI), a metric that measures the similarity between the true cluster labels (labels) and the predicted clusters (predicted_clusters)

```python
helpers.return_ari(labels, predicted_clusters)
```
